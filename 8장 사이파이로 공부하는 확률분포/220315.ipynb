### 8.2 베르누이분포와 이항분포

```python
X~Bern(x;μ)
```

```python
Bern(x;μ) =  μ (if x =1), 1-μ(if x= 0)
```

```python
Bern(x;μ) =  μ^x * (1-μ)^(1-x)
```

### 연습문제 8.2.1

```python
x=1 -> μ
x=0 -> 1-μ
```

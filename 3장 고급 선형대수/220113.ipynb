# 좌표와 변환

### 선형종속과 선형독립

선형종속: 벡터의 선형조합이 영벡터가 되도록 하는 스칼라 계수 c1, c2, ...cn이 존재(모든 계수가 0인 경우 제외)<br>
선형독립: 해당 벡터존재 x

### 선형독립과 선형 연립방정식

c1x1 + c2x2 +... + cnxn = Xc

Xc = 0 -> c = 0

### 선형종속인 경우 (다중공선성)

1. 벡터의 개수 > 벡터의 차원 

2. 값이 같거나 서로 실수배 관계인 벡터 존재 

3. 어떤 벡터가 다른 벡터의 선형조합

### 랭크

열랭크: 행렬의 열벡터 중 서로 독립인 열벡터의 최대 개수

행랭크: 행렬의 행벡터 중 서로 독립인 행벡터의 최대 개수

*열랭크 = 행랭크 = 랭크
(rankA <= min(M, N)

```python
x1 = np.array([[1, 3], [2, 4]])
np.linalg.matrix_rank(x1)
```

```python
x2 = np.array([[1, 3, 5], [2, 3, 7]])
np.linalg.matrix_rank(x2)
```

### 풀랭크

랭크가 행 개수와 열 개수 중 작은 값과 같은 경우

rankA = min(M, N)

### 로우-랭크 행렬

n차원 벡터 m개를 이용하여 만들어지는 행렬

m에 따라 랭크-1 행렬, 랭크2- 행렬, ... 랭크 m행렬 (rankA = M)

### 벡터공간과 기저벡터

* 벡터공간: 벡터 n개가 선형독립일 때 이 벡터들을 선형조합하여 만들어지는 모든 벡터의 집합
* 기저벡터: 벡터공간을 구성하는 벡터공간<br>
벡터공간의 차원 = 기저벡터의 개수

### 랭크와 역행렬

정방행렬이 풀랭크다 <-> 역행렬이 존재한다

### 벡터공간 투영

m개의 n차원 기저벡터 v1, v2, ...vm이 존재할 때(m<n),<br>
모든 n차원 벡터 x에 대해<br>
기저벡터 v1, v2, ... vm을 선형조합하여 만든 벡터와 원래 벡터 x의 차인 x-x^llv가<br>
모든 기저벡터에 직교하면,<br>
그 벡터를 v1, v2, ... vm 벡터공간에 대한  투영벡터라 한다.<br>
이 때 차이벡터 x-x^llv = x^ㅗ를 벡터공간에 대한 직교벡터라 한다.

### 정규직교인 기저벡터로 이루어진 벡터공간

* 만약 기저벡터 v1, v2, ... vm이 정규직교이면 투영벡터 x^ll는 각 기저벡터에 대한 내적값으로 표현됨<br>
x^llv = (x.T@v1)v1 + (x.T@v2)v2 + ... + (x.T@vm)vm<br>
<br>
* 투영벡터 길이의 제곱은 각 기저벡터와의 내적의 제곱합임<br>
(llx^llvll)^ 2 = (x.T@v1)^2 + (x.T@v2)^ 2 + ... + (x.T@vm)^ 2<br>
<br>
* 이 때 직교벡터는 기저벡터들로 이루어진 벡터공간의 모든 벡터에 대해 직교함 <br>
(즉, 벡터 x의 투영벡터는 기저벡터들로 이루어진 벡터공간의 모든 벡터 중에서 가장 벡터 x와 가까운 벡터임.)

### 표준기저벡터

기저벡터이면서 원소 중 하나만 값이 1이고 다른 값은 0으로 이루어진 벡터

### 좌표

기저벡터를 선형조합하여 그 벡터를 나타내기 위한 계수벡터

### 변환행렬

기존 기저벡터 {e1, e2}와 새로운 기저벡터 {g1, g2} 간에,<br>
g1 = 1/2 * e1 + 1/2 * e2<br>
g2 = -1/2 * e1 + 1/2 * e2 라는 관계가 성립할 때,<br>
<br>
e1, e2에 대한 g1, g2의 좌표벡터는 열벡터로 나타낼 수 있다.<br>
<br>
이 때 변환행렬 A = [1/2  -1/2<br> 
                    1/2   1/2]이다. 

### 좌표변환

새로운 기저벡터에 대해 좌표를  계산하는 것<br>
<br>
새로운 기저벡터에 대한 벡터 A의 좌표 = xg = T * xe<br>
* T: 변환행렬 A의 역행렬

```python
e1 = np.array([1, 0])
e2 = np.array([0, 1])
x = np.array([2, 2])
g1 = np.array([1, 1]) / np.sqrt(2)
g2 = np.array([-1, 1]) / np.sqrt(2)
A = np.vstack([g1, g2]). T
A
```

```python
Ainv = np.linalg.inv(A)
Ainv
```

```python
Ainv.dot(x)
```

### 이미지변환

* scipy.ndimage.affine_transform(image_data, A:변환행렬)
* 이 때 표준기저벡터로 e1 = [0, -1], e2 = [1, 0] 사용
* x1이 아래를 향하는 세로축, x2가 오른쪽을 향하는 가로축
* 원점은 왼쪽 상단의 점

```python
import scipy as sp
import scipy.misc
import scipy.ndimage

f = sp.misc.face(gray=True)

e1 = np.array([0, 1]) # matplotlib의 이미지 좌표규약으로 부호변경 
e2 = np.array([1, 0])
E = np.vstack([e1, e2]).T
g1 = np.array([1, 1]) / np.sqrt(2)
g2 = np.array([-1, 1]) / np.sqrt(2)
A = np.vstack([g1, g2]).T
gc1 = E @ g1
gc2 = E @ g2

plt.subplot(121)
plt.imshow(f, cmap=mpl.cm.bone, alpha=0.9)
plt.annotate('', xy=500*e1, xytext=(0,0), arrowprops=green)
plt.annotate('$e_1$', xy=500*e1, xytext=500*e1 + [-100,0])
plt.annotate('', xy=500*e2, xytext=(0,0), arrowprops=green)
plt.annotate('$e_2$', xy=500*e2, xytext=500*e2 + [0, -50])
plt.annotate('', xy=500*gc1, xytext=(0, 0), arrowprops=red)
plt.annotate('$g_1$', xy=500*gc1, xytext=500*gc1 + [50, -50])
plt.annotate('', xy=500*gc2, xytext=(0, 0), arrowprops=red)
plt.annotate('$g_2$', xy=500*gc2, xytext=500*gc2 + [50, 0])
plt.axis("off")
plt.xlim(-200, 1000)
plt.ylim(800, -500)
plt.title("좌표변환전")

f1 = sp.ndimage.affine_transform(f, A)

plt.subplot(122)
plt.imshow(f1, cmap=mpl.cm.bone, alpha=0.8)
plt.annotate('', xy=500*e1, xytext=(0,0), arrowprops=red)
plt.annotate('$g_1$', xy=500*e1, xytext=500*e1 + [-100,0])
plt.annotate('', xy=500*e2, xytext=(0,0), arrowprops=red)
plt.annotate('$g_2$', xy=500*e2, xytext=500*e2 + [0, -50])
plt.axis("off")
plt.xlim(-200, 1000)
plt.ylim(800, -500)
plt.title("좌표변환후")
plt.show()
```

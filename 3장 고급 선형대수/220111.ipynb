### 스칼라와 벡터의 곱

* 양의 실수 X 벡터: 방향 변화x, 실수 크기만큼 벡터 길이 커짐<br>
* 음의 실수 X 벡터: 방향 변화o(반대), 실수 크기만큼 벡터 길이 커짐

### 단위벡터

* 길이가 1인 벡터 (x / np.linalg.norm(x))

### 벡터의 합

* 두 벡터의 합도 벡터<br>
* 그 결과는 두 벡터를 이웃으로 가지는 평행사변형의 대각선 벡터<br>
* 또는 벡터를 더하고자 하는 벡터의 끝점으로 평행이동했을 때 이동한 벡터가 가리키는 점의 위치<br>

### 코사인 유사도

* 두 벡터 사이의 각의 코사인 값
* x.T @ y / llxll x llyl
* 코사인 거리: 1 - 코사인 유사도

### 연습문제 3.1.6

```python
a = np.array([4, 5, 2, 2])
b = np.array([4, 0, 2, 0])
c = np.array([2, 2, 0, 1])

import itertools

score = [a, b, c]
name = ['a', 'b', 'c']
n = itertools.combinations(name, 2)
name_ = list(n)

score_combination = itertools.combinations(score, 2)
score_combination_list = list(score_combination)

cosine_distance = [0] * len(score_combination_list)
for i in range(len(score_combination_list)):
    cosine_distance[i] = name_[i][0] + name_[i][1], 1 - (score_combination_list[i][0] @ score_combination_list[i][1]) / (np.linalg.norm(score_combination_list[i][0]) * np.linalg.norm(score_combination_list[i][1]))

Euclidean_distance = [0] * len(score_combination_list)
for i in range(len(score_combination_list)):
    Euclidean_distance[i] = name_[i][0] + name_[i][1], np.linalg.norm(score_combination_list[i][0] - score_combination_list[i][1])

print('코사인 거리: ', cosine_distance)
print('유클리드 거리: ', Euclidean_distance)
```

### 벡터의 분해와 성분

* 어떤 두 벡터 A, B의 합이 다른 벡터 C가 될 떄 C가 두 벡터 성분 A, B로 분해된다고 한다.

### 투영성분과 직교성분

벡터 a를 다른 벡터 b에 직교하는 성분과 벡터 b에 평행한 성분으로 분해할 수 있다. 
* 이 때 평행한 성분은 투영성분이고,
* 직교인 성분은 직교성분이다.

* 투영성분의 길이 = a.T @ b/llbll
* 투영성분 = 투영성분 길이 x 벡터 b방향의 단위벡터
* 직교성분 = 원래 벡터 - 투영성분벡터

```python
import numpy as np
import matplotlib.pylab as plt

plt.rc("font", size=18)  # 그림의 폰트 크기를 18로 고정
gray = {"facecolor": "gray"}
black = {"facecolor": "black"}
red = {"facecolor": "red"}
green = {"facecolor": "green"}
blue = {"facecolor": "blue"}

a = np.array([2, 2])
b = np.array([-2, 2])
a2 = (a @ b) / np.linalg.norm(b) ** 2 * b
a1 = a - a2
plt.annotate('', xy=b, xytext=(0, 0), arrowprops=green)
plt.annotate('', xy=a2, xytext=(0, 0), arrowprops=blue)
plt.annotate('', xy=a1, xytext=(0, 0), arrowprops=blue)
plt.annotate('', xy=a, xytext=(0, 0), arrowprops=red)
plt.plot(0, 0, 'kP', ms=10)
plt.plot(a[0], a[1], 'ro', ms=10)
plt.plot(b[0], b[1], 'ro', ms=10)
plt.xticks(np.arange(-10, 10))
plt.yticks(np.arange(-10, 10))
plt.xlim(-1.2, 4.1)
plt.ylim(-0.5, 3.2)
plt.show()
```

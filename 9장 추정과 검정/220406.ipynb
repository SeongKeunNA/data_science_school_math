### 연습문제 9.5.2

1.

```python
import seaborn as sns
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
female_smoker = tips[tips.sex == 'Female'].smoker.value_counts() 
female_smoker
```

```python
sp.stats.binom_test(54,87,0.5,'greater')
```

유의확률인 0.015가 유의 수준인 0.1보다 낮으므로 여성 비흡연자와 여성 흡연자의 비율이 같다는 귀무가설을 기각한다.
즉, 여자 손님 중 비흡연자가 흡연자보다 많다고 할 수 있다.

2.

```python
dinner_female_smoker = tips[(tips.sex == 'Female') & (tips.time == 'Dinner')].smoker.value_counts() 
dinner_female_smoker
```

```python
sp.stats.binom_test(29,52,0.5,'greater')
```

유의확률인 0.24가 유의 수준인 0.1보다 높으므로 여성 비흡연자와 여성 흡연자의 비율이 같다는 귀무가설을 기각할 수 없다..
즉, 저녁 시간 대 여자 손님 중 비흡연자가 흡연자보다 많다고 할 수 없다.

### 연습문제 9.5.3

```python
for n in range(1,1500):
    value = sp.stats.binom_test(int(round(n*0.7)),n,2/3,'greater')
    if value < 0.01:
        print('n:', n, 'value:', value)
    break
```

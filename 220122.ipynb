### 정적분

**독립변수 x가 어떤 구간 [a,b] 사이일 때 그 구간에서 함수 f(x)의 값과 수평선 (x축)이 이루는 면적**(을 구하는 행위)

∫a~b f(x) dx

정적분 풀이

1. 심파이 등으로 부정적분 뒤 미적분학 기본 정리 사용
2. 함수의 면적 부분을 실제로 잘개 쪼개어 면적을 근사하게 구하는 **수치미분** 사용

### **미적분학의 기본정리**

∫a~b f(x) dx = F(b) - F(a)

1번 방법 예제 (**subs(), evalf()** 메서드 사용)

```python
x, y = sympy.symbols('x y')
f = x ** 3 - 3 * x ** 2 + x + 6
f
```

```python
# 부정적분
F = sympy.integrate(f)
F
```

```python
(F.subs(x, 2) - F.subs(x, 0)).evalf()
```

### 다변수 정적분

두 변수로 모두 적분하는 것은 2차원 평면에서 주어진 사각형 영역 아래의 부피를 구하는 것과 같다.

a,b : x의 하한과 상한

gfun, hfun: y의 상한과 하한 (x의 함수이어야 함)

```python
def f(x, y):
    return np.exp(-x * y) / y**2

sp.integrate.dblquad(f, 1, np.inf, lambda x:0, lambda x: np.inf)
```

#### 그레디언트 벡터의 특성

1. 그레디언트 벡터의 크기는 기울기를 의미
2. 그레디언트 벡터의 방향은 함수 곡면의 기울기가 가장 큰 방향(단위 길이당 함숫값이 가장 크게 증가하는 방향)
3. 그레디언트 벡터의 방향은 등고선 방향과 직교

테일러 전개식에 의해, f(x) - f(x0) = Δf ~ ∇f(x0).T (x-x0)

즉, 변화의 방향 x-x0가 그레디언트 벡터와 같은 방향일 때 Δf가 가장 커진다.

같은 등고선 위의 다른 점 x1을 향해 움직이는 등고선 방향의 움직임은 x1 - x0이고  f(x0) = f(x1)

따라서 테일러 전개식에 의해, f(x1) - f(x0) = Δf ~ ∇f(x0).T (x1-x0) = 0이므로,

등고성 방향 x1 - x0와  Δf(x0)이 직교한다는 것을 알 수 있다.

### 행렬미분법칙 1: 선형 모형

선형 모형을 미분하면 그레디언트 벡터는 가중치 벡터임.

f(x) = w.Tx

∇f = ∂w.T @ x / ∂ x = ∂x.t @ w / ∂x = ∂(w1x1 + w2x2 + ... + wnwn) / ∂x1 , ∂(w1x1 + w2x2 + ... + wnxn) / ∂x2 , ...<br>
, ∂(w1x1 + x2w2 + ... + wnwn) / ∂xn = w1, w2 , ... wn = **w**

### 행렬미분법칙 2: 이차형식

이차형식을 미분하면 행렬과 벡터의 곱으로 나뉜다.

f(x) = x.T @ A @ x

∇f(x) = ∂x.T @ A @ x / ∂x = **(A+A.T) @ x**

### 벡터를 벡터로 미분하는 경우 

미분 당하는 원소(N)와 미분하는 원소(M)가 여러 개이므로 도함수는 행렬이 됨.(MxN)

### 행렬미분법칙3: 행렬과 벡터의 곱의 미분

행렬 A와 벡터 x의 곱 Ax를 벡터 x로 미분하면 행렬 A.T가 된다.

f(x) = Ax

∇f(x) = ∂(Ax) / ∂x = A.T

**자코비안 행렬**: 벡터함수를 벡터변수로 미분한 결과로 생기는 도함수 행렬의 **전치행렬**

Jf(x) = J = (∂f / ∂x).T

* 그레디언트 벡터의 자코비안 행렬의 전치행렬

Hf(x) = H = J(∇f(x)).T

Hij = ∂^2f / ∂xi∂xj

### 스칼라를 행렬로 미분

### 행렬미분법칙4: 행렬 곱의 대각성분

두 정방행렬을 곱해서 만들어진 행렬의 대각성분은 스칼라이고, 이를 뒤의 행렬로 미분하면 앞의 전치행렬이 나온다.

f(X) = tr(WX)

∂f / ∂X = ∂tr(WX) / ∂X = W.T

### 행렬미분법칙5: 행렬식의 로그

행렬식과 행렬식의 로그 값은 스칼라이다. 이 값을 원래의 행렬로 미분하면 원래 행렬의 역행렬의 전치행렬이 된다.

f(X) = loglXl

∂f / ∂X = ∂loglXl / ∂X = x^-1.T

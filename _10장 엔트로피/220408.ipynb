```python
P0 = np.linspace(0.001, 1-0.001, 1000)
P1 = 1 - P0
H = -P0 * np.log2(P0) - P1 * np.log(P1)

plt.plot(P1, H, '-', label = '엔트로피')
plt.legend()
plt.xlabel('P(Y=1)')
plt.show()
```

```python
import sympy
import math
from scipy import integrate
```

```python
sympy.init_printing(use_latex = 'mathjax')

x, s = sympy.symbols('x s')
p = (1 / sympy.sqrt(2 * sympy.pi * (s**2))) * sympy.exp(-(x**2) / (2 * (s**2)))
f = -p * sympy.log(p)
f
```

```python
def f(x):
    s = 1
    return -(1 / sympy.sqrt(2 * sympy.pi * (s**2))) * sympy.exp(-(x**2) / (2 * (s**2))) * sympy.log((1 / sympy.sqrt(2 * sympy.pi * (s**2))) * sympy.exp(-(x**2) / (2 * (s**2))))
```

```python
integrate.quad(f, -np.inf, np.inf,)
```
